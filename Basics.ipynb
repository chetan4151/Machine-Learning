```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title('sin wave')
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)
plt.subplot(3,1,1)
plt.plot(x,y_sin,'m--')
plt.title('sine')
plt.subplot(3,1,2)
plt.plot(x,y_cos,'r--')
plt.title('cosine')
plt.subplot(3,1,3)
plt.plot(x,y_tan,'g--')
plt.title('tan')
plt.savefig('trigo.pdf')
```
